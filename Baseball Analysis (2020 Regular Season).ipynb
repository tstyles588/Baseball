```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
braves = pd.read_csv('Braves.csv')
dodgers = pd.read_csv('Dodgers.csv')
```

```python
braves
```

```python
braves.head(10).plot(kind='scatter', x='Name', y='AVG', figsize=(16, 8))
plt.title('Starting Line-up Batting Averages (Los Bravos)')
plt.xlabel('Names')
plt.ylabel('AVG')
```

```python
Braves = braves.query('PA > 0')
Braves
```

```python
Stealers= Braves.groupby('Name').SB.sum().nlargest()
Stealers
```

```python
Stealers.plot(figsize= (15, 8), kind='bar')
plt.title('Stolen  Base Leaders (Braves)')
plt.xlabel('Names')
plt.ylabel('Number of Stolen Bases')
```

```python
Hitters = Braves.groupby('Name').ISO.sum().nlargest()
Hitters
```

```python
Duds = Braves.groupby('Name').K.sum().nlargest()
Duds
```

```python
braves.head(10).plot(kind='scatter', x='Name', y='wRC', figsize=(16, 8))
plt.title('Most Runs Created (Braves Starting Line-up)')
plt.xlabel('Names')
plt.ylabel('wRC')
```

```python
Dodgers = dodgers.query('PA > 0')
Dodgers
```

```python
Stealerz = Dodgers.groupby('Name').SB.sum().nlargest()
Stealerz
```

```python
Hitterz = Dodgers.groupby('Name').ISO.sum().nlargest()
Hitterz.plot(kind='bar', figsize = (15, 8))
plt.title('Hardest Hitters on the Dodgers (by ISO)')
plt.xlabel('Names')
plt.ylabel('ISO')
```

```python
Dudz = Dodgers.groupby('Name').K.sum().nlargest()
Dudz
```

```python
dodgers.head(10).plot(kind='scatter', x='Name', y='AVG', figsize=(16, 8))
plt.title('Starting Line-up Batting Averages')
plt.xlabel('Names')
plt.ylabel('AVG')
```
